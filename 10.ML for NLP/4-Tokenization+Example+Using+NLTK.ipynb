```python
# paragraph
corpus="""Hello Welcome,to Krish Naik's NLP Tutorials.
Please do watch the entire course! to become expert in NLP.
"""
```

```python
print(corpus)
```

```python
import nltk
nltk.download('punkt_tab')
```

```python
##  Tokenization
## Sentence-->paragraphs
from nltk.tokenize import sent_tokenize
```

```python
documents=sent_tokenize(corpus)
```

```python
documents
```

```python
type(documents)
```

```python
for sentence in documents:
    print(sentence)
```

```python
## Tokenization 
## Paragraph-->words
## sentence--->words
from nltk.tokenize import word_tokenize
```

```python
word_tokenize(corpus)
```

```python
for sentence in documents:
    print(word_tokenize(sentence))
```

```python
# A simple regex-based tokenizer that splits text into words and punctuation based on word boundaries.
# It separates all non-word characters (e.g., punctuation) into separate tokens
from nltk.tokenize import wordpunct_tokenize
```

```python
# Doesn't account for linguistic nuances (e.g., contractions like "don't" are split into ['don', "'", 't']).
# Lacks context-aware tokenization.
```

```python
wordpunct_tokenize(corpus)
```

```python
from nltk.tokenize import TreebankWordTokenizer
```

```python
tokenizer=TreebankWordTokenizer()
```

```python
tokenizer.tokenize(corpus)
```
